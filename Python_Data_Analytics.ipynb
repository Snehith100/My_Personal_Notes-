```python
# Pandas:

''' Important Pandas Concepts to Master
1. Data Structures

Series (1D labeled array)

DataFrame (2D table with rows & columns)

2. Data Input & Output

pd.read_csv(), pd.to_csv() (most common)

Also: read_excel(), read_sql(), read_json()

3. Indexing & Selection

loc[] → label-based selection

iloc[] → position-based selection

Boolean filtering (df[df["col"] > 10])

Setting index (set_index, reset_index)

4. Data Cleaning & Preprocessing

Handling missing data → isna(), fillna(), dropna()

Removing duplicates → drop_duplicates()

Renaming columns → rename()

Changing data types → astype()

5. Exploratory Analysis

head(), tail(), info(), describe()

value_counts()

unique(), nunique()

6. Data Transformation

Apply functions → apply(), map(), applymap()

Column operations (df["new"] = df["A"] + df["B"])

String operations (str.lower(), str.contains())

7. Sorting & Ranking

sort_values()

sort_index()

rank()

8. GroupBy & Aggregation

groupby() for grouping

Aggregations → sum(), mean(), count(), agg()

Pivot tables (pivot_table())

9. Merging & Joining

concat() (stacking DataFrames)

merge() (SQL-style joins)

join() (joining by index)

10. Reshaping

melt() (long format)

pivot() (wide format)

stack() / unstack()

11. Time Series Handling

pd.to_datetime()

Resampling (resample("M"))

Date indexing & filtering

12. Performance Tricks

Vectorization (avoid loops)

df.query() for fast filtering

Using categorical data for memory efficiency'''
```

```python
# Pandas Contin...

# It is used to analyze data and one of the important library for working with datasets
# It is for analyzing, cleaning, exploring and manipulating data
# It is called as "Panel Data" or "Python Data Analysis".

# We can import pandas like this
import pandas as pd # if not installed just write # install pandas

# Pandas Data Structures
# I-Dimensional Labelled array

a = pd.Series([1,2,3], index=["a","b","c"])
a

# pd.series will create one dimensional labelled array

# 2-Dimensional Labelled array (2-D):

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Salary": [50000, 60000, 70000]
}

df = pd.DataFrame(data)
print(df)

# pd.DataFrame will create table with rows and columns
#---------------------------------------------

# Reading and Writing Data

# Read CSV file
df = pd.read_csv("file.csv")
# loads data from a CSV file into a DataFrame.

# Save DataFrame to CSV
df.to_csv("output.csv", index=False)
# Loads data into dataframe from CSV file.

#----------------------------------------------

# Exploring Data

print(df.head())      # first 5 rows
print(df.tail(3))     # last 3 rows
print(df.shape)       # (rows, columns)
print(df.info())      # column info
print(df.describe())  # summary stats for numbers

# ------------------------------------------

# Selecting Data column and Row section

print(df["Name"])       # single column
print(df[["Name","Age"]])  # multiple columns

# Rows:

print(df.iloc[0])       # first row (position-based)
print(df.loc[0])        # first row (label-based index)

#-------------------------------------------------------

# Filtering Rows

print(df[df["Age"] > 30])   # rows where Age > 30

#-----------------------------------------------------

# Indexing

df2 = df.set_index("Name")   # make Name the index
print(df2.loc["Alice"])      # select by index label

#-------------------------------------------------

# Cleaning Data

# Add missing values
df["Bonus"] = [1000, None, 1500]

print(df.isna())        # check missing
df["Bonus"] = df["Bonus"].fillna(0)  # replace missing with 0

# Rename columns
df.rename(columns={"Salary": "Income"}, inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Change data type
df["Age"] = df["Age"].astype(float)

#----------------------------------------------------


# Summary and Counting

print(df["Department"].value_counts())  # frequency count
print(df["Department"].unique())        # unique values
print(df["Department"].nunique())       # number of unique values

# ----------------------------------------------

# Soring and Ranking

print(df.sort_values(by="Age"))      # sort by Age
print(df["Income"].rank())           # rank incomes

#------------------------------------------

# Creating New columns and applying Math operations

df["Income_K"] = df["Income"] / 1000       # math operation
df["Age+10"] = df["Age"].apply(lambda x: x+10)  # apply function

#---------------------------------------------

# String Operations

df["Dept_lower"] = df["Department"].str.lower()
print(df["Department"].str.contains("IT"))   # check text


#---------------------------------------------------

# Grouping and Aggregation

print(df.groupby("Department")["Income"].mean())   # average per dept

print(df.groupby("Department")["Income"].agg(["mean","max"]))

#----------------------------------------------------

# Pivot Tables

print(df.pivot_table(values="Income", index="Department", aggfunc="mean"))

#------------------------------------------------------

# Merging and Joining

extra = pd.DataFrame({
    "Department": ["HR", "IT"],
    "Manager": ["Tom", "Jerry"]
})

merged = pd.merge(df, extra, on="Department")
print(merged) # Same like SQL


#-------------------------------------------------------

# Reshaping Data

# Wide to long
print(pd.melt(df, id_vars=["Name"], value_vars=["Age","Income"]))

# Pivot (long → wide)
print(df.pivot(index="Name", columns="Department", values="Income"))

#------------------------------------------------------------

# Time series

df["Joining_Date"] = pd.to_datetime(["2020-01-01", "2019-06-15", "2018-03-10"])
print(df["Joining_Date"].dt.year)  # extract year

ts = pd.Series([100,200,300],
               index=pd.date_range("2022-01-01", periods=3, freq="M"))
print(ts.resample("M").mean())     # resample monthly

#---------------------------------------------------------

# Performance tricks in pandas

# Vectorization (fast, no loops)
df["Double_Income"] = df["Income"] * 2

# Query syntax
print(df.query("Age > 30 and Department == 'IT'"))

# Convert to category (saves memory)
df["Department"] = df["Department"].astype("category")


#----------------------------------------------------------------
```

```python
# Scipy continuation:

# Import the libraries
import numpy as np
from scipy import stats # take a part of scipy dont need whole library and
                        # importing statistical module from scipy

#-----------------------------------------------------

# Descriptive statistics (scipy.stats)

import numpy as np
from scipy import stats

data = [10, 20, 20, 40, 50, 50, 50, 60]

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data, keepdims=True)) # keepdims=True → keeps result in array form for consistency.
print("Standard Deviation:", np.std(data))

#------------------------------------------------------------

# Probability distributions

# Probabilities tell us how likely an event is. Normal distribution is commonly used in data analysis.

from scipy.stats import norm # Import normal distribution functions

mean = 0
std_dev = 1 # Spread of distribution how wide curve is

prob = norm.cdf(1, loc=mean, scale=std_dev) # cummulative distributive fxn gives value <=1
print("Probability:", prob)

#--------------------------------------------------------------
# Linear Algerbra

# Used to solve equations and work with matrices, important in regression and ML.

import numpy as np
from scipy import linalg

A = np.array([[2, 1],
              [1, -1]])  # coefficient matrix
b = np.array([5, 1])      # result vector

x = linalg.solve(A, b)
print("Solution:", x)

#----------------------------------------------------------------

# Optimization

# Finding minimum or maximum of a function (common in ML, cost functions).

from scipy.optimize import minimize # optimize means finding best value (lowest/highest)

def f(x):
    return (x - 3)**2

result = minimize(f, x0=0) # minimum value it is inside optimize
print("Best x:", result.x)

#-------------------------------------------------------------

# Integration (scipy. integration)
# Integration = area under a curve, used in probability, totals, cumulative sums.

from scipy import integrate

f = lambda x: x**2 # labda creates little fxn f(x) = x2
area, error = integrate.quad(f, 0, 3) # quad is tool inside integrate finds exact value under curve.
print("Area:", area)

#------------------------------------------------------


# Distances (scipy.spatial)
# Distance = how far points are, used in clustering and similarity.

from scipy.spatial import distance # distance is straight line distance

p1 = [1, 2]
p2 = [4, 6]

dist = distance.euclidean(p1, p2) # euclidean  = sqrt of (x2-x1)2 - (y2-y1)2
print("Distance:", dist)

#---------------------------------------------------------

# Fourier Tranform

# Convert time-domain data → frequency-domain, used in signal/time series analysis.

import numpy as np
from scipy import fft

x = np.array([1, 2, 3, 4])
y = fft.fft(x)
print("Fourier Transform:", y)
```

```python
# Matplotlib:
# Matplotlib’s main module is pyplot, usually imported as plt.

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]   # X-axis values
y = [2, 4, 6, 8, 10]  # Y-axis values

# Plotting
plt.plot(x, y)        # line plot
plt.title("Basic Line Plot")  # title of chart
plt.xlabel("X Axis")          # label for x-axis
plt.ylabel("Y Axis")          # label for y-axis
plt.show()            # display the plot

#---------------------------------------------------------

# Scatter plot (Relationship between two variables):

x = [5, 7, 8, 7, 6, 9, 5, 3, 4, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="red", marker="o") # marker means dot shape
plt.title("Scatter Plot Example")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.show()

#---------------------------------------------------

# Bar chart (Categorical Comparision)

categories = ["Apples", "Bananas", "Cherries"]
values = [10, 15, 7]

plt.bar(categories, values, color="green")
plt.title("Bar Chart Example")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()

#-------------------------------------------------------

# Histogram (Distribution of data )

import numpy as np

data = np.random.randn(1000)  # 1000 random numbers these numbers usually stay close to 0, but can go negative or positive
                              # (because it’s a "normal distribution" — like test scores where most people are average, few are very high/low).

plt.hist(data, bins=30, color="skyblue", edgecolor="black") # cutting number line os 30 boxes. # bins are like boxes
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()

#-------------------------------------------

# Multiple plots in pone figure

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label="Squares")
plt.plot(x, y2, label="Line")
plt.legend()  # show labels means displays which box is which basicalaly shows legend box
plt.title("Multiple Lines")
plt.show()

#----------------------------------------------

# Subplots (Multiple charts side by side)

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.subplot(1, 2, 1)   # Here (1,2,1) and (1,2,2) means 1st row and 2nd column beacuase we have two col names with one
                       # i.e (line plot (1st col) and Bar chart (2nd col)) and Line plot in first postion and bar chart in
                       # second position. So (1,2) is common for both where postion is differed
plt.plot(x, y)
plt.title("Line Plot")

plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.title("Bar Chart")

plt.show()

#-----------------------------------------------

# Pie chart (Proportions)

sizes = [40, 30, 20, 10]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90) # sizes means shows size and label names their names
                                                                # %1.1f%% means 1 digit before and after decimal
plt.title("Pie Chart Example")  # startangle means rotate circle by 90 degrees without it rotates from 3'o clock
plt.show()

#--------------------------------------------------

# Styling and customization

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color="purple", linestyle="--", marker="s", linewidth=2)
plt.title("Styled Line")
plt.show()


```
